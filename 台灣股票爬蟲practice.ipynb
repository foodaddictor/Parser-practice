```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
re = requests.get("https://tw.stock.yahoo.com/d/i/rank.php?t=pri&e=tse")
```

```python
print(re.status_code)
```

```python
soup = BeautifulSoup(re.text, "html.parser")
```

```python
th = soup.find_all("th")
```

```python
d= []
for i in th:
    d.append(i.text)
d[3] = "漲跌"
d[4] = "漲跌幅"
print(d)
```

```python
tr = soup.find_all("tr",{"align":"right"})
```

```python

stock = {}
for i in range(10):
    a_i = []
    for j in tr:
        td = j.find_all("td")
        a_i.append(td[i].text)
    stock[d[i]] = a_i

#for loop 注意誰先誰後，選錯就做不出來了
```

```python
import pandas as pd
```

```python
TW_yahoo_stock = pd.DataFrame(stock, columns = stock.keys()) #加columns = dict.keys() df 欄位才不會隨機，才會照著順序
```
